```python
from one.api import ONE
one = ONE(base_url='https://openalyx.internationalbrainlab.org', password='international', silent=True)
one.list_datasets()
```

```python
eids, info = one.search(dataset='probes.description', task_protocol='_iblrig_tasks_ephysChoiceWorld6.4.2', details=True)
```

```python
print(eids)
```

```python
dats=[]
for eid in eids:
  try:
    dat = one.load_dataset(eid, '*channels.brainLocationIds_ccf_2017.npy')
    dats.append(dat)
  except:
    print('this eid doesn\'t have brain loc acr: ', eid)
```

```python
dats
```

```python
import numpy as np
unique=[]
for d in dats:
  u=list(np.unique(d))
  for j in u:
    unique.append(j)
unique=list(np.unique(unique))
```

```python
print(unique)
```

```python
len(unique)
```

```python
brain_hist=np.zeros((len(unique)))
dats_un=[]
for d in dats:
  u=list(np.unique(d))
  dats_un.append(u)

for u in range(len(unique)):
  for d in range(len(dats_un)):
      if unique[u] in dats_un[d]:
        brain_hist[u]+=1

import matplotlib.pyplot as plt

plt.bar(range(0,len(unique)),np.array(brain_hist))

      
```
